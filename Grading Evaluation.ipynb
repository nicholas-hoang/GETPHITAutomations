```python
def get_files():
    path = pathlib.Path('.')
    files = []
    
    for folder in path.iterdir():
        if folder.is_dir():
            for file in folder.iterdir():
                if file.is_file() and file.suffix == '.xlsx':
                    files.append(file)
    return files
```

```python
files = get_files()
```

```python
# def write_scores(files) is a function that takes in a list of files and reads the data from each file.
# It appends the data to the existing data frame and finally returns the full excel file.

def write_scores(files):
    # Create empty data frame 
    student_evals = pd.DataFrame(columns=['Student Name', 'Grader 1', 'Grader 2', 'Score 1', 'Score 2'])

    for file in files:
        # read data from file
        data = pd.read_excel(file)
        
        try: 
            student_name = data.iloc[1,1]
            grader_1 = data.iloc[4,0]
            grader_2 = data.iloc[5,0]

            # Check if there are enough rows and columns in the data frame
            if data.shape[0] >= 5 and data.shape[1] >= 8:
                score_1 = data.iloc[4,7]
                score_2 = data.iloc[5,7]
                average = (score_1 + score_2) / 2
                new_row = pd.DataFrame({'Student Name': [student_name], 'Grader 1': [grader_1], 'Grader 2': [grader_2], 'Score 1': [score_1], 'Score 2': [score_2], 'Average': [average]})
                student_evals = pd.concat([student_evals, new_row], ignore_index=True)
        
        except IndexError:
            continue 

    # Write the data frame to Excel
    student_evals.to_excel('student_evals.xlsx', index=False)

    # return the full Excel file
    return pd.read_excel('student_evals.xlsx')

write_scores(files)
```
